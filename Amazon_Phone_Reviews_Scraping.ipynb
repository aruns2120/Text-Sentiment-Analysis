```python
import requests
from bs4 import BeautifulSoup
```

```python
search_query = 'mobile+phones+all+price'
base_url = "https://www.amazon.in/s?k="
url = base_url+search_query
url
```

```python
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'}
search_response = requests.get(url, headers = header)
search_response.status_code
```

## Utility functions

```python
cookie = {} # insert request cookies within{}
def getAmazonSearch(search_query):
    '''A function to get the content of the page of given query'''
    url = "https://www.amazon.in/s?k="+search_query
    print(url)
    page = requests.get(url, cookies = cookie, headers = header)
    if page.status_code == 200:
        return page
    else:
        return "Error"
```

```python
def Searchasin(asin):
    '''A function to get the contents of individual product pages using ASIN (Amazon Standard Identification Number)'''
    url = "https://www.amazon.in/dp/"+asin
    print(url)
    page = requests.get(url,cookies = cookie, headers = header)
    if page.status_code == 200:
        return page
    else:
        return "Error"
```

```python
def Searchreviews(review_link):
    '''This function takes the link of 'see all reviews' and extracts the content'''
    url = "https://www.amazon.in"+review_link
    print(url)
    page = requests.get(url, cookies = cookie, headers = header)
    if page.status_code == 200:
        return page
    else:
        return "Error"
```

## Extracting Reviews

```python
from tqdm import tqdm

product_names = []
reviews = []
# We need to extarct reviews of 100 mobile phones.
# Each page gives 16 ASINs on an average, so 7 pages will give 112 ASINs which would satisfy our requirement

for i in tqdm(range(1,8)): # Number of seaarch result pages to look for (1 to 7)
                        
    data_asin = []
    response = getAmazonSearch(search_query +'&page=' + str(i))
    soup = BeautifulSoup(response.content)
    for asin in soup.findAll("div", {'class':"sg-col-20-of-24 s-result-item s-asin sg-col-0-of-12 sg-col-28-of-32 sg-col-16-of-20 sg-col sg-col-32-of-36 sg-col-12-of-16 sg-col-24-of-28"}):
        #"div", {'class':"sg-col-20-of-24 s-result-item s-asin sg-col-0-of-12 sg-col-28-of-32 sg-col-16-of-20 sg-col sg-col-32-of-36 sg-col-12-of-16 sg-col-24-of-28"} tag is for asin

        data_asin.append(asin['data-asin'])

    link = []
    for asin in range(len(data_asin)):
        response = Searchasin(data_asin[asin])
        soup = BeautifulSoup(response.content)
        for l in soup.findAll("a",{'data-hook':"see-all-reviews-link-foot"}):
            link.append(l['href'])
    
    for l in range(len(link)):
        for p in range(10): # Number of pages to extract reviews from
        # Each page contains 10 reviews. Looking at first 10 pages should give 100 reviews per phone
            response = Searchreviews(link[l]+'&pageNumber='+str(p))
            soup = BeautifulSoup(response.content)
            for name in soup.findAll("a", {'data-hook': "product-link", 'class' : "a-link-normal"}):
                
                for review in soup.findAll("span",{'data-hook':"review-body"}):
                    product_names.append(name.text) #adding the product names to the list
                    reviews.append(review.text)
    
```

## Save extracted data to .txt file

```python
import pandas as pd
df = pd.DataFrame({'product_name' : product_names, 'review' : reviews})#converting this dictionary into a dataframe
print(df.shape)
df.head(10)
```

```python
df['review'] = df['review'].apply(lambda x : x.replace('\\n', '').strip())
```
